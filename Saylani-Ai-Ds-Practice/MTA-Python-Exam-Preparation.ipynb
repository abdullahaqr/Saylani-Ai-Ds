## Q35

```python
inventory = open("inventory.txt", 'r')
eof = False
while eof == False:
    line = inventory.readline()
    if line != '':
        if line != "\n":
            print(line)
    else:
        print("End of file")
        eof = True
        inventory.close()
```

## Q39

```python
def print_table(file):
    data = open(file, 'r')
    for record in data:
        fields = record.split(",")
        print("{0:10} {1:5.1f} {2:7.2f}".format(fields[0], eval(fields[1]), eval(fields[2])))
print_table('fruits.txt')
```
